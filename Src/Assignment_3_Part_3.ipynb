#### Doing all the imports in start

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Downloading and putting the DataSet from birghtspace (D2L) to local directory and loading it using pandas

```python
# Load the dataset
house_price_data = pd.read_csv('../data/train.csv')
```

#### Inspecting the dataset to understand its structure and contents.

#### Check for missing values, data types, and summary statistics.

Display the first few rows

```python
house_price_data.head()
```

 Check for size and different datatypes  of the columns

```python
house_price_data.info()
```

Summary statistics

```python
house_price_data.describe()
```

setting Id as index and removing it as a column since we already have 80 columns to deal with

```python
house_price_data = house_price_data.set_index(house_price_data.Id).drop('Id', axis=1)
```

#### Checking for missing values

```python
missing_values = house_price_data.isnull().sum()
missing_values[missing_values > 0].sort_values(ascending=False)
```

#### Visualize missing values

```python
plt.figure(figsize=(12, 6))
sns.heatmap(house_price_data.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

#### Distribution Analysis of the Target Variable (SalePrice) 

```python
sns.histplot(house_price_data['SalePrice'], kde=True, bins=30, color='skyblue')
plt.title("Distribution of SalePrice")
plt.ylabel("Frequency")
plt.show()
```

#### Since Often SalePrice is skewed; we will apply log transformation for a more normal shape

```python
house_price_data['LogSalePrice'] = np.log1p(house_price_data['SalePrice'])
sns.histplot(house_price_data['LogSalePrice'], kde=True, bins=30)
plt.title('Distribution of Log-Transformed SalePrice')
plt.xlabel('Log(SalePrice)')
plt.ylabel('Frequency')
plt.show()
```

#### --- Correlation Analysis ---
#### Compute correlation matrix for numerical features

```python
numerical_cols = house_price_data.select_dtypes(include=[np.number]).columns
corr_matrix = house_price_data[numerical_cols].corr()
```

#### Plot heatmap of the correlation matrix


```python
plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=False)
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```

#### Identify top features correlated with SalePrice

```python
top_corr = corr_matrix['SalePrice'].sort_values(ascending=False)
print("\nTop features correlated with SalePrice:")
display(top_corr.head(10))
```

#### Scatter Plots for Top Correlated Features 
#### Exclude SalePrice itself and select top 5 features

```python
top_features = top_corr.index[1:6]
plt.figure(figsize=(15,10))
for i, feature in enumerate(top_features):
    plt.subplot(2, 3, i+1)
    sns.scatterplot(x=house_price_data[feature], y=house_price_data['SalePrice'])
    plt.title(f'SalePrice vs {feature}')
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
plt.tight_layout()
plt.show()
```

#### Distribution of Numeric Features

```python
numeric_features = house_price_data.select_dtypes(include=[np.number]).columns
house_price_data[numeric_features].hist(bins=30, figsize=(20,15))
plt.tight_layout()
plt.show()
```

#### Boxplot for a Key Categorical Variable
#### For example, OverallQual (Overall Material and Finish Quality)

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='OverallQual', y='SalePrice', data=house_price_data, palette="Set3", hue="OverallQual", legend=False) #used hue and legend to get rid of deprecation warning
plt.title("SalePrice by Overall Quality")
plt.xlabel("Overall Quality")
plt.ylabel("SalePrice")
plt.show()
```

#### Identify potential outliers

```python
outliers = house_price_data[(house_price_data['GrLivArea'] > 4000) & (house_price_data['SalePrice'] < 300000)]
print(outliers[['GrLivArea', 'SalePrice']])
```

#### Feature engineering: Create 'HouseAge' feature

```python
house_price_data['HouseAge'] = 2025 - house_price_data['YearBuilt']
plt.figure(figsize=(10, 6))
sns.scatterplot(x=house_price_data['HouseAge'], y=house_price_data['SalePrice'])
plt.title('HouseAge vs SalePrice')
plt.xlabel('HouseAge')
plt.ylabel('SalePrice')
plt.show()
```

#### Feature engineering: Create 'HouseAge' feature

Creating the HouseSqFt feature by adding TotalBsmtSF and GrLivArea

```python
house_price_data['HouseSqFt'] = house_price_data['TotalBsmtSF'] + house_price_data['GrLivArea']
```

Plotting HouseSqFt vs SalePrice

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=house_price_data['HouseSqFt'], y=house_price_data['SalePrice'])
plt.title('HouseSqFt vs SalePrice')
plt.xlabel('HouseSqFt')
plt.ylabel('SalePrice')
plt.show()
```

#### Feature engineering: Create 'TotalBathrooms' feature

```python
house_price_data['TotalBathrooms'] = house_price_data['BsmtFullBath'] + 0.5 * house_price_data['BsmtHalfBath'] + house_price_data['FullBath'] + 0.5 * house_price_data['HalfBath']
plt.figure(figsize=(10, 6))
sns.scatterplot(x=house_price_data['TotalBathrooms'], y=house_price_data['SalePrice'])
plt.title('TotalBathrooms vs SalePrice')
plt.xlabel('TotalBathrooms')
plt.ylabel('SalePrice')
plt.show()
```
